```python
def print_even_odd(start,end):
    """ This function does even odd calculation within a range
    """
    for i in range(start,end+1):
        if (i%2 == 0):    # Checked remainder == 0, To check even or odd
            print(f"{i} is even")
        else:
            print(f"{i} is odd")

print_even_odd(10,20)
```

```python
# # Python Numbers

# Python supports several types of numbers:
# - Integers (int): Whole numbers, e.g., 42, -7
# - Floating-point numbers (float): Numbers with decimals, e.g., 3.14, -0.001
# - Complex numbers (complex): Numbers with real and imaginary parts, e.g., 2 + 3j

# Examples:
int_num = 42           # Integer
float_num = 3.14       # Float
complex_num = 2 + 3j   # Complex

print("Integer:", int_num, type(int_num))
print("Float:", float_num, type(float_num))
print("Complex:", complex_num, type(complex_num))

# You can perform arithmetic operations:
print("Addition:", int_num + float_num)
print("Multiplication:", int_num * 2)
print("Complex addition:", complex_num + (1 + 1j))

# Python automatically converts between types when needed
result = int_num + float_num
print("Result type after addition:", type(result))
```

```python
# # Python Operators Overview

# Operators are special symbols used to perform operations on variables and values.

x, y, z = 4, 1, 3
c = 10
# ## 1. Arithmetic Operators
print("Addition:", x + y)         # Adds x and y
print("Subtraction:", x - y)      # Subtracts y from x
print("Multiplication:", x * z)   # Multiplies x and z
print("Division:", c / x)         # Divides c by x
print("Modulus:", c % x)          # Remainder of c divided by x
print("Exponentiation:", x ** z)  # x raised to the power z
print("Floor Division:", c // x)  # c divided by x, rounded down

print("-" * 40)  # Separator
# ## 2. Comparison Operators
print("x == y:", x == y)          # Checks if x equals y
print("x != y:", x != y)          # Checks if x not equals y
print("x > y:", x > y)            # Checks if x greater than y
print("x < y:", x < y)            # Checks if x less than y
print("x >= y:", x >= y)          # Checks if x greater or equal to y
print("x <= y:", x <= y)          # Checks if x less or equal to y

print("-" * 40)  # Separator
# ## 3. Logical Operators
bool_literal = True
print("bool_literal and (x > y):", bool_literal and (x > y))  # True if both are True
print("bool_literal or (x < y):", bool_literal or (x < y))    # True if any is True
print("not bool_literal:", not bool_literal)                  # Negates bool_literal

print("-" * 40)  # Separator
# ## 4. Assignment Operators
temp = 5
temp += 2   # Adds 2 to temp
print("temp after += 2:", temp)

print("-" * 40)  # Separator
# ## 5. Membership Operators
a = [2,3,4]
print("2 in a:", 2 in a)          # Checks if 2 is in list a
#print("'ok' in anotherValidName:", 'ok' in anotherValidName)  # Checks substring in string

print("-" * 40)  # Separator
# ## 6. Identity Operators
a,b = 400,400
print("a is b:", a is b)          # Checks if a and b refer to same object
print("a is b:", a is b)          # Checks if a and b refer to same object

print("-" * 40)  # Separator

print("x is y:", x is y)          # Checks if x and y refer to same object

print("-" * 40)  # Separator

# ## 7. Bitwise Operators
print("x & z:", x & z)            # Bitwise AND
print("x | z:", x | z)            # Bitwise OR
print("x ^ z:", x ^ z)            # Bitwise XOR
print("~x:", ~x)                  # Bitwise NOT
print("x << 1:", x << 1)          # Bitwise left shift
print("z >> 1:", z >> 1)          # Bitwise right shift
```

```python
# Sequence data types in depth: List, String, Tuple
# Each section demonstrates creation, common operations, and important differences.

# -------------------- LIST (mutable, ordered) --------------------
print("\n--- LIST ---")
lst = [1, 2, 3, "py", [10, 20]]
print("original list:", lst)

# Indexing & slicing
print("lst[0]:", lst[0])
print("last element (negative index):", lst[-1])
print("slice (1:4):", lst[1:4])
print("slice with step (::-1) reversed:", lst[::-1])

# Mutation methods (in-place)
lst.append(4)
lst.extend([5, 6])
lst.insert(1, 1.5)
print("after append/extend/insert:", lst)

# Remove/pop
val = lst.pop()           # removes last
lst.remove(1.5)           # removes first occurrence
print("after pop/remove:", lst, "popped value:", val)

# In-place vs creating new: sort() vs sorted()
num_list = [3, 1, 4, 2]
sorted_copy = sorted(num_list)   # returns new list
num_list.sort()                  # sorts in-place
print("sorted_copy:", sorted_copy, "num_list after sort():", num_list)

# Aliasing and shallow copy
a = [[1], [2]]
b = a           # alias, modifications reflect in both
c = a.copy()    # shallow copy: top-level copied, inner lists shared
a[0].append(99)
print("alias b:", b, "shallow copy c:", c)

# List comprehensions & iteration
squares = [x*x for x in range(6) if x % 2 == 0]
print("comprehension (even squares):", squares)
for i, v in enumerate(["a", "b", "c"]):
    print("enumerate:", i, v)

# Unpacking
first, second, *rest = [10, 20, 30, 40]
print("unpack:", first, second, rest)
```

```python
c# -------------------- STRING (immutable, ordered sequence of characters) --------------------
print("\n--- STRING ---")
s = "  Hello, World!  "
print("original:", repr(s))

# Indexing & slicing
print("s[0]:", s[0])
print("slice [2:9]:", s[2:9])
print("upper/lower:", s.upper(), s.lower())
print("strip():", s.strip())

# Immutability: attempting to assign raises TypeError (commented out to allow cell execution)
# s[0] = 'h'  # TypeError: 'str' object does not support item assignment

# Common string operations
parts = s.split(",")
print("split by comma:", parts)
joined = " & ".join(["a", "b", "c"])
print("join example:", joined)

# Search and replace
print("find 'World':", s.find("World"))
print("replace:", s.replace("World", "Universe"))

# Formatting & f-strings
name = "Alice"
age = 30
print(f"{name} is {age} years old")

# Encoding/decoding (to bytes)
b = s.encode("utf-8")
print("encoded bytes:", b)
print("decoded back:", b.decode("utf-8"))
```

```python
# -------------------- TUPLE (immutable, ordered) --------------------
print("\n--- TUPLE ---")
t = (1, 2, 3)
single = (5,)          # single-element tuple needs trailing comma
empty = ()
print("tuple t:", t, "single:", single, "empty:", empty)

# Indexing & slicing
print("t[1]:", t[1])
print("t[:2]:", t[:2])

# Immutability: cannot assign to elements (commented to avoid error)
# t[0] = 10  # TypeError: 'tuple' object does not support item assignment

# Packing & unpacking
p = 10, 20, 30          # packing without parentheses
x, y, z = p             # unpacking
print("packed p:", p, "unpacked x,y,z:", x, y, z)

# Useful methods (tuples are smaller API than lists)
print("count of 2 in p:", p.count(2))
print("index of 20 in p:", p.index(20))

# Tuples are hashable if all their elements are hashable -> usable as dict keys
d = {}
coords = (10, 20)
d[coords] = "point A"
print("dict with tuple key:", d)

# Concatenation and repetition produce new tuples
t2 = t + (4, 5)
print("concatenated tuple:", t2)
print("repeated tuple:", t * 2)
```
