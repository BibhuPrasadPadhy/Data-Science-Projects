```python
import numpy as np
```

```python
def forward_pass(x, w, b):
    z = np.dot(w, x) + b
    return z

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict(x, w, b):
    z = forward_pass(x, w, b)
    a = sigmoid(z)
    if a >= 0.5:
        return 1
    else:
        return 0    
```

```python
x = np.array([0.5, 0.2, 0.1])
w = np.array([0.4, 0.3, 0.2]) #Assume Learned weights
b = 0.1 #Assume Learned bias

predict(x,w,b)
```
