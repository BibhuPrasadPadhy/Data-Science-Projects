```python
class Car:
    def __init__(self, color, make="Toyota"):
        self.make = make
        self.color = color
        self._top_speed = 300
        self.__engine_number = "ENG12345"


    def drive(self):
        print(f"Car is driving a car of {self.make} make and {self.color} color")

    def get_engine_number(self):
        return self.__engine_number
    
    def get_top_speed(self):
        return self._top_speed

car1 = Car("Red")
car1.drive()
car1._top_speed = 350
print(f"Top speed of the car is: {car1.get_top_speed()} km/h")
# Accessing private attribute using name mangling
print(f"Engine number of the car is: {car1._Car__engine_number}")
```

```python
#Abstraction 
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= amount
    def get_balance(self):
        return self.__balance

chaitanya_account = BankAccount(1000)
chaitanya_account.deposit(500)

bibhu_account = BankAccount(2000)
bibhu_account.deposit(1000)

chaitanya_account.withdraw(200)
bibhu_account.withdraw(500) 

print(f"Chaitanya's account balance: {chaitanya_account.get_balance()}")
print(f"Bibhu's account balance: {bibhu_account.get_balance()}")
```

```python
class Human:
    def __init__(self, name, age,   gender, occupation, is_married, children):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation
        self.is_married = is_married
        self.children = []
    
    def speak(self):
        print("Human speaks")
    def walk(self):
        print("Human walks")
    def eat(self):
        print("Human eats")
    def sleep(self):
        print("Human sleeps")
    def work(self):
        print("Human works")


class Girl(Human):
    def __init__(self, name, age, gender,occupation, is_married, children,hobbies,use_myntra):
        super().__init__(name, age, "Female", occupation,is_married, children)
        self.hobbies = hobbies
        self.use_myntra = use_myntra
    def speak(self):
        print("Girl speaks  softly  and sweetly")

    def cook(self):
        print("Girl cooks delicious meals")
    
chaitanya = Girl("Chaitanya", 25, "Female", "Software Engineer", False, [], ["Cooking", "Traveling"], True)
chaitanya.speak()

ramesh = Human("Ramesh", 5, "Male", "Engineer", False, [])
ramesh.speak()
```

### Imputation - Null Value Replacement - Null value Imputation

```python
## Step 1 : Identify why the data is missing, is it missing at random or not, and then decide how to handle the missing data.
## Step 2: If the data is missing at random, we can impute the missing values using mean, median, or mode.
## Step 3: If the data is not missing at random, we can either drop the rows with missing values or use advanced imputation techniques like similar value imputation/replacement.
## Note : If the amount of missing data is too much > 20%, then first check with client/stakeholder if you can drop that column, if yes then drop it, if no, Then look for row dropping or imputation using source check, similar case/value replacement
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['City'].fillna(df['City'].mode()[0], inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
df['Designation'].fillna(df['Designation'].mode()[0], inplace=True)
df['Experience_Years'].fillna(df['Experience_Years'].mean(), inplace=True)
df['Joining_Date'].fillna(df['Joining_Date'].mode()[0], inplace=True)
df['Performance_Rating'].fillna(df['Performance_Rating'].mean(), inplace=True)
```
