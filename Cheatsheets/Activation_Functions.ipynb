```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def softmax(z):
    z_adjusted = z - np.max(z, axis=1, keepdims=True)
    exp_z = np.exp(z_adjusted)
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def relu(z):
    return np.maximum(0, z)

def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha * z, z)

def tanh(z):
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

def swish(z):
    return z * sigmoid(z)

def gelu(z):
    return 0.5 * z * (1 + np.tanh(np.sqrt(2 / np.pi) * (z + 0.044715 * np.power(z, 3))))    
```

```python
z = np.array([[-1, 2, 3], [-4, 45, 6]])
print("Sigmoid:\n", sigmoid(z))
print("Softmax:\n", softmax(z))
print("ReLU:\n", relu(z))  
print("Leaky ReLU:\n", leaky_relu(z))
print("Tanh:\n", tanh(z))
print("Swish:\n", swish(z))
print("GELU:\n", gelu(z))
```
